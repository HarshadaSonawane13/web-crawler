```python
# Import necessary libraries
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
```

```python
# Step 1: Target site (demo e-commerce site for scraping practice)
url = "https://books.toscrape.com/catalogue/page-1.html"
```

```python
# Step 2: Fetch the webpage content
response = requests.get(url)

# Use .content (better for full HTML data)
soup = BeautifulSoup(response.content, "html.parser")

print("✅ Webpage fetched and parsed successfully!")
```

```python
# Step 3: Find all product containers on the page
products = soup.find_all("article", class_="product_pod")

print(f"✅ Found {len(products)} products on the page.\n")
```

```python
# Step 4: Extract product name, price, and link
print("Product Information from BooksToScrape:\n")

for product in products:
    name = product.h3.a["title"]                      # Product name
    price = product.find("p", class_="price_color").text  # Product price
    link = product.h3.a["href"]                       # Product link (relative)
    full_link = urljoin("https://books.toscrape.com/catalogue/", link)  # Absolute URL

    print(f"Name : {name}")
    print(f"Price: {price}")
    print(f"Link : {full_link}")
    print("-" * 60)
```
